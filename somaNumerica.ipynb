```python
import numpy as np
```

# Parâmetros

```python
A = 10.0 ** np.arange(1, 16)  # A = 10^1, 10^2, ..., 10^15
```

# Cálculo e saída

```python
for Ai in A:
    sum_result = Ai
    for k in range(int(1e7)):
        sum_result += 1e-7
    error = abs(sum_result - (Ai + 1))
    print(f'A = {Ai:.0e}, Soma = {sum_result:.16f}, Erro = {error:.2e}')
```
