# Método da bisseção para f1(x) = ln(x) + 2x e f2(x) = exp(-x) - sin(x)
# lista do aluno, da apostila do prof. Castilho!

```python
import numpy as np
import matplotlib.pyplot as plt
```


# Cores Math-Dynamics

```python
neonpink = '#FF007F'
goldlove = '#FFD700'
financeblue = '#0E1A40'
```

# Função bisseção

```python
def bisection(f, a, b, tol):
    if not np.isscalar(a) or not np.isscalar(b) or not np.isscalar(tol):
        raise ValueError('Erro: a, b, e tol devem ser escalares.')
    while (b - a) / 2 > tol * abs((a + b) / 2):
        c = (a + b) / 2
        fc = f(c)
        if not np.isscalar(fc):
            raise ValueError('Erro: f(c) deve retornar um valor escalar.')
        if abs(fc) < 1e-10:
            return c
        elif f(a) * fc < 0:
            b = c
        else:
            a = c
    return (a + b) / 2
```

# Funções

```python
f1 = lambda x: np.log(x) + 2 * x  # ln(x) + 2x
f2 = lambda x: np.exp(-x) - np.sin(x)  # exp(-x) - sin(x)
```


# Parâmetros

```python
a1, b1, tol = 0.01, 0.5, 0.1  # Intervalo para f1
a2, b2 = 0, 1  # Intervalo para f2
```


# Verificação para f1

```python
if a1 <= 0 or b1 <= 0:
    raise ValueError('Erro: Intervalo de f1 deve ter x > 0 para ln(x).')
if f1(a1) * f1(b1) >= 0:
    raise ValueError(f'Erro: f1(a1) e f1(b1) devem ter sinais opostos (f1(a1)={f1(a1):.4f}, f1(b1)={f1(b1):.4f}).')
```


# Verificação para f2

```python
if f2(a2) * f2(b2) >= 0:
    raise ValueError(f'Erro: f2(a2) e f2(b2) devem ter sinais opostos (f2(a2)={f2(a2):.4f}, f2(b2)={f2(b2):.4f}).')
```

# Chamar bisseção

```python
x1 = bisection(f1, a1, b1, tol)
x2 = bisection(f2, a2, b2, tol)
```

# Exibir resultados

```python
print(f'Raiz a) de f1(x) = ln(x) + 2x: {x1:.4f}')
print(f'Raiz b) de f2(x) = exp(-x) - sin(x): {x2:.4f}')
```

# Plotar f1

```python
x_f1 = np.linspace(0.01, 0.5, 100)
y_f1 = f1(x_f1)
plt.figure(figsize=(6, 4), facecolor=financeblue)
plt.plot(x_f1, y_f1, color=neonpink, linewidth=2)
plt.plot(x1, f1(x1), 'o', color=goldlove, markersize=8)
plt.axhline(0, color='black', linestyle='--')
plt.xlabel('x', color='white')
plt.ylabel('f_1(x)', color='white')
plt.title('Raiz de ln(x) + 2x', color='white')
plt.grid(True)
plt.gca().set_facecolor(financeblue)
plt.gca().spines['bottom'].set_color('white')
plt.gca().spines['left'].set_color('white')
plt.gca().tick_params(colors='white')
plt.savefig('f1_root.png', dpi=300, bbox_inches='tight', facecolor=financeblue)
plt.show()

```

# Plotar f2

```python
x_f2 = np.linspace(0, 1, 100)
y_f2 = f2(x_f2)
plt.figure(figsize=(6, 4), facecolor=financeblue)
plt.plot(x_f2, y_f2, color=neonpink, linewidth=2)
plt.plot(x2, f2(x2), 'o', color=goldlove, markersize=8)
plt.axhline(0, color='black', linestyle='--')
plt.xlabel('x', color='white')
plt.ylabel('f_2(x)', color='white')
plt.title('Raiz de exp(-x) - sin(x)', color='white')
plt.grid(True)
plt.gca().set_facecolor(financeblue)
plt.gca().spines['bottom'].set_color('white')
plt.gca().spines['left'].set_color('white')
plt.gca().tick_params(colors='white')
plt.savefig('f2_root.png', dpi=300, bbox_inches='tight', facecolor=financeblue)
plt.show()
```

```python
def fixed_point(g, x0, tol, max_iter):
    x = x0
    for i in range(max_iter):
        x_new = g(x)
        if abs(x_new - x) / abs(x_new) < tol:
            break
        x = x_new
    return x_new
```

# newton_raphson.py
# Método de Newton-Raphson para f4(x) = 2cos(x) - e^{x/2} e f6(x) = (5-x)e^x - 1
# lista do aluno, da apostila do prof. Castilho!

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Cores Math-Dynamics

```python
neonpink = '#FF007F'
goldlove = '#FFD700'
financeblue = '#0E1A40'
```

# Função newton_raphson

```python
def newton_raphson(f, df, x0, tol, max_iter):
    x = x0
    for i in range(max_iter):
        x_new = x - f(x) / df(x)
        if abs(x_new - x) / abs(x_new) < tol:
            break
        x = x_new
    return x_new
```


# Funções e derivadas

```python
f4 = lambda x: 2 * np.cos(x) - np.exp(x / 2)
df4 = lambda x: -2 * np.sin(x) - 0.5 * np.exp(x / 2)
f6 = lambda x: (5 - x) * np.exp(x) - 1
df6 = lambda x: (4 - x) * np.exp(x)
```


# Executar

```python
x4 = newton_raphson(f4, df4, 0.4, 1e-3, 100)
x6 = newton_raphson(f6, df6, 4.75, 1e-3, 100)
```

# Exibir resultados

```python
print(f'Raiz d): {x4:.4f}')
print(f'Raiz f): {x6:.4f}')
```

# newton_pth_root.py
# Método de Newton-Raphson para raiz p-ésima de x^p = a

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Cores Math-Dynamics

```python
neonpink = '#FF007F'
goldlove = '#FFD700'
financeblue = '#0E1A40'
```

# Função newton_pth_root

```python
def newton_pth_root(a, p, x0, tol, max_iter):
    x = x0
    for i in range(max_iter):
        x_new = (1 / p) * ((p - 1) * x + a / (x ** (p - 1)))
        if abs(x_new - x) / abs(x_new) < tol:
            break
        x = x_new
    return x_new
```

# Parâmetros

```python
a, p = 16, 4  # Exemplo: quarta raiz de 16
x = newton_pth_root(a, p, 2, 1e-3, 100)
```


# Exibir resultado

```python
print(f'Raiz {p}-ésima de {a}: {x:.4f}')
```

# Função original para plotagem

```python
f = lambda x: x**p - a
x_f = np.linspace(1, 3, 100)
y_f = f(x_f)
plt.figure(figsize=(6, 4), facecolor=financeblue)
plt.plot(x_f, y_f, color=neonpink, linewidth=2)
plt.plot(x, f(x), 'o', color=goldlove, markersize=8)
plt.axhline(0, color='black', linestyle='--')
plt.xlabel('x', color='white')
plt.ylabel(f'$x^{p} - {a}$', color='white')
plt.title(f'Raiz {p}-ésima de {a}', color='white')
plt.grid(True)
plt.gca().set_facecolor(financeblue)
plt.gca().spines['bottom'].set_color('white')
plt.gca().spines['left'].set_color('white')
plt.gca().tick_params(colors='white')
plt.savefig('pth_root.png', dpi=300, bbox_inches='tight', facecolor=financeblue)
plt.show()
```

# newton_sincos.py
# Método de Newton-Raphson para f(x) = sin(cos(sqrt(3)*x))

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
neonpink = '#FF007F'
goldlove = '#FFD700'
financeblue = '#0E1A40'
```

# Função newton_raphson

```python
def newton_raphson(f, df, x0, tol, max_iter):
    x = x0
    for i in range(max_iter):
        x_new = x - f(x) / df(x)
        if abs(x_new - x) < tol:
            break
        x = x_new
    return x_new
```

# Função e derivada

```python
f = lambda x: np.sin(np.cos(np.sqrt(3) * x))
df = lambda x: np.cos(np.cos(np.sqrt(3) * x)) * (-np.sin(np.sqrt(3) * x)) * np.sqrt(3)
```
