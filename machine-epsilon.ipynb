```python
import numpy as np
```

# Parâmetros

```python
A = [1, 10, 100, 1000]
```


# Cálculo e saída

```python
for Ai in A:
    Ep = 1.0
    while Ai + Ep > Ai:
        Ep = Ep / 2
    print(f'A = {Ai:.0f}, Epsilon = {2*Ep:.2e}')
```
