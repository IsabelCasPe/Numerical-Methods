# Aproximação da derivada de f(x) = sin(x) em x = pi e x = 0
# lista do aluno, da apostila do prof. Castilho!

```python
import numpy as np
```

# Funções

```python
f = lambda x: np.sin(x)  # f(x) = sin(x)
df_exact = lambda x: np.cos(x)  # f'(x) = cos(x)
```


# Parâmetros

```python
x = [np.pi, 0]  # Pontos de avaliação
h = 10.0 ** np.arange(-1, -11, -1)  # h = 10^-1, 10^-2, ..., 10^-10
```

# Cálculo e saída

```python
for xi in x:
    print(f'x = {xi:.2f}, f\'(x) exato = {df_exact(xi):.4f}')
    for hj in h:
        df_approx = (f(xi + hj) - f(xi)) / hj
        error = abs(df_exact(xi) - df_approx)
        print(f'h = {hj:.0e}, f\'(x) aprox = {df_approx:.6f}, erro = {error:.2e}')
```

# derivative_approx_with_plots.py
# Aproximação da derivada de f(x) = sin(x) em x = pi e x = 0 com gráficos


```python
import numpy as np
import matplotlib.pyplot as plt
```


# Cores Math-Dynamics

```python
neonpink = '#FF007F'
goldlove = '#FFD700'
financeblue = '#0E1A40'
```


# Funções

```python
f = lambda x: np.sin(x)  # f(x) = sin(x)
df_exact = lambda x: np.cos(x)  # f'(x) = cos(x)
```

# Parâmetros

```python
x = [np.pi, 0]  # Pontos de avaliação
h = 10.0 ** np.arange(-1, -11, -1)  # h = 10^-1, 10^-2, ..., 10^-10
```


# Cálculo e gráficos

```python
for i, xi in enumerate(x):
    errors = []
    print(f'x = {xi:.2f}, f\'(x) exato = {df_exact(xi):.4f}')
    for hj in h:
        df_approx = (f(xi + hj) - f(xi)) / hj
        error = abs(df_exact(xi) - df_approx)
        errors.append(error)
        print(f'h = {hj:.0e}, f\'(x) aprox = {df_approx:.6f}, erro = {error:.2e}')

    # Plotar erro vs h
    plt.figure(figsize=(6, 4), facecolor=financeblue)
    plt.loglog(h, errors, color=neonpink, linewidth=2, label='Erro')
    plt.scatter(h, errors, color=goldlove, s=50, label='Pontos')
    plt.xlabel('h', color='white')
    plt.ylabel('Erro', color='white')
    plt.title(f'Erro na Aproximação de f\'({xi:.2f})', color='white')
    plt.grid(True, which="both", ls="--")
    plt.legend()
    plt.gca().set_facecolor(financeblue)
    plt.gca().spines['bottom'].set_color('white')
    plt.gca().spines['left'].set_color('white')
    plt.gca().tick_params(colors='white')
    plt.savefig(f'derivative_error_x{xi:.2f}.png', dpi=300, bbox_inches='tight', facecolor=financeblue)
    plt.show()

print("Gráficos salvos: derivative_error_x3.14.png, derivative_error_x0.00.png")
```

 # Plotar erro vs h

```python
plt.figure(figsize=(6, 4), facecolor=financeblue)
plt.loglog(h, errors, color=neonpink, linewidth=2, label='Erro')
plt.scatter(h, errors, color=goldlove, s=50, label='Pontos')
plt.xlabel('h', color='white')
plt.ylabel('Erro', color='white')
plt.title(f'Erro na Aproximação de f\'({xi:.2f})', color='white')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.gca().set_facecolor(financeblue)
plt.gca().spines['bottom'].set_color('white')
plt.gca().spines['left'].set_color('white')
plt.gca().tick_params(colors='white')
plt.savefig(f'derivative_error_x{xi:.2f}.png', dpi=300, bbox_inches='tight', facecolor=financeblue)
plt.show()

print("Gráficos salvos: derivative_error_x3.14.png, derivative_error_x0.00.png")
```
